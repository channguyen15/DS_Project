### Mỗi cột có ý nghĩa gì?

Theo file mô tả dữ liệu thì: 
- <code>Invoice id</code>: số hóa đơn 
- <code>Branch</code>: Ký hiệu chi nhánh của siêu thị (A: nằm ở thành phố Yangon, B: nằm ở thành phố Mandalay, C: nằm ở thành phố Naypyitaw)
- <code>City</code>: Tên thành phố của 3 chi nhánh siêu thị 
- <code>Customer type</code>: Phân loại khách hàng (có thẻ thành viên hoặc không có thẻ thành viên)
- <code>Gender</code>: Giới tính của khách hàng 
- <code>Product line</code>: Phân loại mặt hàng của sản phẩm 
- <code>Unit price</code>: Giá của 1 sản phẩm 
- <code>Quantity</code>: số lượng sản phẩm 
- <code>Tax</code>: 5% phí thuế khi mua hàng 
- <code>Total</code>: Tổng giá đã bao gồm thuế
- <code>Date</code>: Ngày mua hàng (từ 1/2019 đến 3/2019)
- <code>Time</code>: Giờ mua hàng
- <code>Payment</code>: Phương thức thanh toán (tiền mặt, thẻ tín dụng và Ewallet)
- <code>COGS</code>: Giá vốn của đơn hàng
- <code>Gross margin percentage</code>: Tỷ lệ phần trăm của lợi nhuận gộp
- <code>Gross income</code>: doanh thu của 1 đơn hàng
- <code>Rating</code>: Đánh giá của khách hàng về trải nghiệm mua sắm (từ 1 đến 10)

**Tiền xử lí**  

Ta thấy:   
- Cột <code>Branch</code>, <code>City</code> tương đương nhau
- Cột <code>Gross margin percentage</code> có ý nghĩa bao quát hơn cột <code>Gross income</code> và có thể tình được từ cột <code>Total</code>, <code>COGS</code>   
    $\rightarrow$ Xóa 2 cột <code>Branch</code>, <code>Gross income</code>  
    ⇒ Vậy nên cần tiền xử lí trước khi tiếp tục khám phá dữ liệu  

```python
columns = ['Branch', 'gross income']
df = df.drop(columns, axis=1)
df
```

### Mỗi cột hiện đang có kiểu dữ liệu gì? Có cột nào có kiểu dữ liệu chưa phù hợp để có thể xử lý tiếp không?

```python
df.dtypes
```

Có 2 cột có có kiểu dữ liệu không phù hợp để xử lý tiếp:   
- Cột <code>Date</code> sang kiểu datetime
- Cột <code>Time</code> sang kiểu datetime.time

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M').dt.time
df.dtypes
```

Hiện tại, các cột có vẻ đã có kiểu dữ liệu phù hợp để xử lý tiếp

### Với mỗi cột có kiểu dữ liệu dạng numeric, các giá trị được phân bố như thế nào?
Phần trăm các giá trị thiếu là bao nhiêu?  
Min? max? Có gì bất thường không?

```python
def missing_ratio (col):
    return col.isna().mean()*100

num_cols = df.select_dtypes(include=np.number)
num_cols_profiles = num_cols.agg([missing_ratio, pd.Series.min, pd.Series.max])
num_cols_profiles
```

Các cột đều không có dữ liệu bị thiếu   
Cột <code>Unit price</code>, <code>Quantity</code>, <code>Tax 5%</code>, <code>Total</code>, <code>cogs</code> không có gì bất thường  
Cột <code>gross margin percentage</code> có min, max bằng nhau điều này nói lên rằng các giá trị trong cột này đều bằng nhau thể hiện siêu thị này đang áp dụng cùng tỷ lệ chi phí và doanh thu cho tất cả các mặt hàng hoặc giao dịch  
Cột <code>Rating</code> đánh giá từ 1 đến 10 nên min, max như vậy là chính xác, không có giá trị thiếu   
$\rightarrow$ Toàn bộ các cột trên không có gì bất thường  

### Với mỗi cột có kiểu dữ liệu dạng categorical, các giá trị được phân bố như thế nào? 
Phần trăm các giá trị thiếu là bao nhiêu?  
Có bao nhiêu giá trị khác nhau? Hiển thị một vài  
Có gì bất thường không?

```python
def diff_vals(s): # danh sách các giá trị khác nhau trong một cột
    return s.dropna().unique()
def num_diff_vals(s): # số lượng giá trị khác nhau trong một cột
    return s.dropna().nunique()
cate_cols = df.select_dtypes(exclude=np.number)
cate_cols_profiles= cate_cols.agg([missing_ratio, num_diff_vals, diff_vals])
cate_cols_profiles
```

Các cột đều không có dữ liệu bị thiếu
Cột <code>Invoice ID</code> có giá trị bằng với số dòng vì mỗi hóa đơn đều có ID riêng biệt không thể bị trùng  
Cột <code>City</code> có ba giá trị trị tương ứng với ba siêu thị  
Cột <code>Customer type</code> có hai giá trị tương ứng với hai loại khách hàng là thành viên hoặc không phải thành viên  
Cột <code>Gender</code> có hai giá trị tương ứng với hai giới tính   
Cột <code>Product line</code> có 6 giá trị khác nhau thể hiện 6 loại mặt hàng của siêu thị   
Cột <code>Date</code>, <code>Time</code> có nhiều giá trị vì có nhiều ngày cũng như nhiều giờ trong ngày  
Cột <code>Payment</code>có ba giá trị trị tương ứng với ba phương thức thanh toán  
$\rightarrow$ Toàn bộ không có gì bất thường

## V. Tiền xử lý + phân tích dữ liệu để trả lời từng câu hỏi

### Câu 1. Xu Hướng chi tiêu giữa khách hàng nam và nữ?

#### Tiền xử lí

Để lấy có thể trả lời được câu hỏi này, ta cần dữ liệu của các cột <code>Gender</code>, <code>Product line</code>, <code>Total</code> 

```python
df1 = df[['Gender', 'Product line', 'Total']]
df1
```

#### Phân tích dữ liệu và trả lời cho câu hỏi
Để có thể trả lời cho câu hỏi, ta sẽ làm như sau:
- Công đoạn 1: Tính toán chi tiêu của nữ cho từng ngành hàng
- Công đoạn 2: Tính toán chi tiêu của nam cho từng ngành hàng
- Công đoạn 3: Từ kết quả ở trên, ta vẽ đồ thị barplot để đánh giá xu hướng chi tiêu giữa nam và nữ

**Công đoạn 1:** Tính toán chi tiêu của nữ cho từng ngành hàng  

```python
female_spending = df1[df1['Gender'] == 'Female']
female_spending_summary = female_spending.groupby('Product line')['Total'].sum().reset_index()
female_spending_summary
```

**Công đoạn 2:** Tính toán chi tiêu của nam cho từng ngành hàng  

```python
male_spending = df1[df1['Gender'] == 'Male']
male_spending_summary = male_spending.groupby('Product line')['Total'].sum().reset_index()
male_spending_summary
```
